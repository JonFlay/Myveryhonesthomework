```python
print("Я хочу поиграть с тобой в игру, дружок...")
print("Введите начало региона (числа, не страны):")
start = int(input())
print("Введите конец региона (всё ещё число):")
finish = int(input())

print("Если бежишь, пиши: exit")

baby = (start + finish) // 2    # середина региона

while True:
    print("Твоё число <", ">", "или = ", baby, "?")
    ans = input()

    if ans == "exit":
        break

    if ans == "=":
        print("АЗАЗАЗАЗА!!:", baby)
        break

    if ans == "<":
        baby = baby - 1

    if ans == ">":
        baby = baby + 1

    if start > finish:
        print("Ты дурачок или что.")
        break
```
