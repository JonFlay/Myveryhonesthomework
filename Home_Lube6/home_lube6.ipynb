```python
numba=int(input())
def fib(numba):
    if numba == 0:
        return 0
    if numba == 1:
        return 1
    return fib(numba-1) + fib(numba-2)

print(f"На улице {numba} стоит тарелка {fib(numba)}.")
```

```python
fib_cache = {}

numba2=int(input())

def fib_cached(numba2):
    if numba2 in fib_cache:
        return fib_cache[numba2]
    if numba2 == 0:
        fib_cache[numba2] = 0
    elif numba2 == 1:
        fib_cache[numba2] = 1
    else:
        fib_cache[numba2] = fib_cached(numba2-1) + fib_cached(numba2-2)
    return fib_cache[numba2]

print(f"На улице {numba2} стоит тарелка {fib_cached(numba2)}.")
```
